```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
df = pd.read_csv('./csv/weight/height_weight.csv',encoding='utf-8',delimiter=',')
```

```python
x = df['身長']
y = df['体重']
```

```python
def basic():
  fig = plt.figure(figsize=(5,3))
  ax = fig.add_subplot(111)
  ax.scatter(x,y)
  ax.set(
      title='身長と体重',
      xlabel='身長',
      ylabel='体重',
  )
  ax.grid()
  plt.show()
```

```python
if __name__ == '__main__':
    basic()
```
