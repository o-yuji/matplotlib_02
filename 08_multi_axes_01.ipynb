```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
x = np.linspace(0,4,5)
y = np.linspace(40,60,5)
```

```python
def showPlot(x,y):
    fig = plt.figure(figsize=(8,4))
    ax1 = fig.add_subplot(2,2,1)
    ax2 = fig.add_subplot(2,2,2)
    ax3 = fig.add_subplot(2,1,2)

    ax1.plot(x,y)
    ax2.bar(x,y)
    ax3.scatter(x,y)

    ax1.set(title='分割1')
    ax2.set(title='分割2')
    ax3.set(title='分割3')

    plt.suptitle('サンプル')
    plt.tight_layout(pad=1,w_pad=1,h_pad=1)

    plt.show()
```

```python
if __name__ == '__main__':
    showPlot(x,y)
```
