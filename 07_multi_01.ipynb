```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import japanize_matplotlib
```

```python
x = np.linspace(-2,5,71)
```

```python
def showMultiPlot():
    fig = plt.figure(figsize=(8,3))
    ax1 = fig.add_subplot(1,2,1)
    ax1.plot(x,np.sin(x),label='sin')
    ax1.plot(x,np.log(x),label='log')
    ax1.plot(x,np.sqrt(x),label='sqrt')
    ax1.plot(x,1/x,label='1/x')
    ax1.plot(x,x*x,label='x^2')
    ax1.set(
        title='複数のグラフ',
        xlabel='値:x',
        ylabel='値:y',
        xlim=(-2,5),
        ylim=(-2.5,2.5),
    )
    ax1.legend(loc='best')
    ax1.grid()
    
    ax2 = fig.add_subplot(1,2,2)
    ax2.plot(x,np.cos(x))

    plt.show()
```

```python
if __name__ == ('__main__'):
    showMultiPlot()
```
