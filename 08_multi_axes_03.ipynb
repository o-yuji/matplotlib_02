```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
df_math = pd.read_csv('./csv/tokuten/tokuten.csv',encoding='utf-8',delimiter=',')
y1 = df_math['得点']
x1 = np.linspace(0,y1.count(),y1.count())
```

```python
df_weight = pd.read_csv('./csv/weight/height_weight.csv')
x2 = df_weight['身長']
y2 = df_weight['体重']
```

```python
Apr = [5,7,6,8,7,9]
May = [0,4,3,7,8,9]
June = [6,7,4,5,6,8]
y3 = [Apr,May,June]
x3 = [1,2,3,4,5,6]
```

```python
y4 = [10,30,20,40,80,20]
x4 = ['SF','ドラマ','スリラー','コメディ','アクション','ロマンス']
```

```python
x=[x1,x2,x3,x4]
y=[y1,y2,y3,y4]
```

```python
def hist(ax):
    ax.hist(y[0],bins=10)
    ax.axvline(x=y[0].mean(),color='r')
    ax.set(
        title='得点分布',
        xlabel='得点',
        ylabel='人数',
    )

def scatter(ax):
    ax.scatter(x[1],y[1])
    ax.set(
        title='身長と体重の分布',
        xlabel='身長',
        ylabel='体重',
    )

def stackplot(ax):
    x3labels = ['Apr','May','June']
    ax.stackplot(x[2],y[2],labels=x3labels)
    ax.set(
        title='プロダクト品質',
        xlabel='原因コード',
        ylabel='発生件数',
        ylim=(0,50),
    )
    ax.legend(loc=2)

def pie(ax):
    ax.pie(
        y[3],
        labels=x[3],
        autopct='%1.1f%%',
        startangle=90,
        counterclock=False,
    )
```

```python
def showPlot(x,y):
    fig = plt.figure(figsize=(7,7))

    gs = fig.add_gridspec(2,2)
    ax1 = fig.add_subplot(gs[0,0])
    ax2 = fig.add_subplot(gs[0,1])
    ax3 = fig.add_subplot(gs[1,0])
    ax4 = fig.add_subplot(gs[1,1])

    hist(ax1)
    scatter(ax2)
    stackplot(ax3)
    pie(ax4)

    plt.tight_layout(pad=4,w_pad=1,h_pad=1)
    plt.suptitle('sample')
    plt.show()
```

```python
if __name__ == '__main__':
    showPlot(x,y)
```
