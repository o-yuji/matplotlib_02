```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
def showFigure():
    fig = plt.figure(figsize=(8,6))
    ax = fig.add_subplot(111)

    bstyle= ['circle','darrow','larrow','rarrow','round','round4','roundtooth','sawtooth','square']
    
    for i ,bstyle in enumerate(bstyle):
        ax.text(
            i%3*3+2,
            int(i/3)*3+2,
            bstyle,
            color='w',
            fontsize=20,
            ha='center',
            bbox={
                'boxstyle':bstyle,
                'fc':'C{}'.format(i),
            },
        )

    ax.set(
        title='BoxStyle',
        xlim=(0,10),
        ylim=(0,10),
    )

    plt.show()
```

```python
if __name__ == '__main__':
    showFigure()
```
