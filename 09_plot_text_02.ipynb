```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
def showFigure():
    fig = plt.figure(figsize=(4,4))
    ax = fig.add_subplot(111)

    bstyle= ['circle','darrow','larrow','rarrow']
    
    for i ,bstyle in enumerate(bstyle):
        ax.text(2,i,'text')

    ax.set(
        title='BoxStyle',
        xlim=(0,10),
        ylim=(0,8),
    )

    plt.show()
```

```python
if __name__ == '__main__':
    showFigure()
```
