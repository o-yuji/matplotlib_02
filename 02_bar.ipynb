```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
month = [str(i) + "月" for i in range(1,13)]
sold = [55,45,50,60,75,90,110,105,100,95,80,70]
```

```python
def intro(x,y):
    fig = plt.figure(figsize=(5,3))
    ax = fig.add_subplot(111)
    ax.set(
        title='アイスコーヒー売上推移',
        xlabel='月度',
        ylabel='売上高(千円)'
    )
    bar_rect = ax.bar(x,y)

    for rect in bar_rect:
        x = rect.get_x()
        y = rect.get_height()
        width = rect.get_width()
        ax.text(x + width/2, 1.002*y,int(y),ha='center',va='bottom')
        # ax.text(x+0.04, 1.005*y,int(y))

    plt.show()
```

```python
if __name__ == '__main__':
    intro(month,sold)
```
