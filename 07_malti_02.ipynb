```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
japanese = [50,60,80,70,65]
math = [90,80,75,50,90]
idx = np.linspace(0,4,5)
```

```python
x = idx
y = [japanese,math]
```

```python
def showPlot(x,y):
    fig = plt.figure(figsize=(8,4))
    ax1 = fig.add_subplot(1,1,1)
    width=0.35
    ax1.bar(x,y[0],width,label='国語',color='lightblue')
    ax1.bar(x+0.4,y[1],width,label='数学',color='orange')
    ax1.set(
        title='国語',
        xlabel='試験回数',
        ylabel='得点',
        xticks=x+width/2,
        xticklabels=[str(i) + '回目' for i in range(1,6)],
        ylim=(0,120)
    )
    plt.legend()
    plt.show()
```

```python
if __name__ == '__main__':
    showPlot(x,y)
```
