```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python

df = pd.read_csv('csv/nikkei/nikkei.csv',header=None,encoding='utf-8',delimiter=',')
```

```python
np_data = np.array(df)
```

```python
x = np_data[:,0]
y = np_data[:,1]
```

```python
def intro(x,y):
    fig = plt.figure(figsize=(10,4))
    ax = fig.add_subplot(111)
    # ax.set_title('日経平均')
    # ax.set_xlabel('期間(2020/3/2-2020/3/31)')
    # ax.set_ylabel('株価(円)')
    ax.set(
        title='日経平均',
        xlabel='期間(2020/3/2-2020/3/31)',
        ylabel='株価(円)',
    )

    ax.plot(x,y)
    plt.show()
```

```python
if __name__ == '__main__':
    intro(x,y)
```
