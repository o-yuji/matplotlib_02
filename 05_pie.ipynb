```python
import pandas as pandas
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
sizes = [10,30,20,40,80,20]
labels = ['SF','ドラマ','スリラー','コメディ','アクション','ロマンス']
```

```python
def showPiePlot():
    fig = plt.figure(figsize=(4,4))
    ax = fig.add_subplot(111)
    explode = [0,0,0,0,0.1,0]
    ax.pie(
        sizes,
        labels=labels,
        counterclock=False,
        autopct='%1.1f%%',
        shadow=True,
        startangle=90,
        explode=explode,
        )
    ax.set_title('好きな映画',fontsize=18)
    plt.show()
```

```python
if __name__ == '__main__':
    showPiePlot()
```
