```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
def showArrow(ax):
    for i in range(1,9):
        ax.arrow(5,i,2 if i%2==0 else -2,1,width=0.1,head_width=0.3,head_length=0.3)
    ax.set(
        title='arrow',
        xlim=(0,10),
        ylim=(0,10),
    )
    ax.scatter([5,5,5,5,5,5,5,5],[1,2,3,4,5,6,7,8])
    ax.grid()
```

```python
def showPlot():
    fig = plt.figure(figsize=(8,4))
    ax1 = fig.add_subplot(111)

    showArrow(ax1)

    plt.show()
```

```python
if __name__ == '__main__':
    showPlot()
```
