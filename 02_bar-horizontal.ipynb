```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
month = [str(i) + "月" for i in range(1,13)]
sold = [55,45,50,60,75,90,110,105,100,95,80,70]
```

```python
def intro(x,y):
    fig = plt.figure(figsize=(7,4))
    ax = fig.add_subplot(111)
    ax.set(
        title='アイスコーヒー売上推移',
        xlabel='売上高(千円)',
        ylabel='月度'
    )
    bar_rect = ax.barh(x,y)

    for rect in bar_rect:
        x = rect.get_width()
        y = rect.get_y()
        height = rect.get_height()
        ax.text(1.002*x, y+height/2, int(x), ha='left', va='center')

    plt.show()
```

```python
if __name__ == '__main__':
    intro(month,sold)
```
