```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
x = [i for i in range(1,7)]
print(x)
```

```python
Apr = [5,7,6,8,7,9]
May = [0,4,3,7,8,9]
June = [6,7,4,5,6,8]
labels = ['Apr','May','June']
```

```python
def showPlt():
  fig = plt.figure(figsize=(5,4,))
  ax = fig.add_subplot(111)
  ax.stackplot(x,Apr,May,June,labels=labels)
  ax.set(
    title='プロダクト品質',
    xlabel='原因コード',
    ylabel='発生件数',
  )
  ax.legend(loc=2)

  plt.show()
```

```python
if __name__ == '__main__':
    showPlt()
```
