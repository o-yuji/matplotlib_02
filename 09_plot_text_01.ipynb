```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
def text1(ax):
    ax.scatter([2,2,2],[8,6,4],color='r')
    ax.grid(True)
    ax.text(2,8,'text')
    ax.text(2,6,'center',ha='center')
    ax.text(2,4,'right',ha='right')
    ax.set(
        title = 'horizontal',
        xlim = (0,4),
        ylim = (0,10),
    )

def text2(ax):
    ax.scatter([2,2,2],[8,6,4],color='r')
    ax.text(2,8,'top',va='top')
    ax.text(2,6,'center',va='center')
    ax.text(2,4,'bottom',va='bottom')
    ax.grid(True)
    ax.set(
        title='vertical',
        xlim = (0,4),
        ylim = (0,10),
    )

def text3(ax):
    ax.grid(True)
    ax.scatter([1,2,3],[2,5,8],color='r')
    ax.text(1,2,'90° bottom',rotation=90,va='bottom')
    ax.text(2,5,'90° center',rotation=90,va='center')
    ax.text(3,8,'90° top',rotation=90,va='top')
    ax.set(
        title='routate',
        xlim = (0,4),
        ylim = (0,10),
    )
```

```python
def plotText():
    fig = plt.figure(figsize=(8,5))
    gs = fig.add_gridspec(2,3)

    ax1 = fig.add_subplot(gs[:,0])
    ax2 = fig.add_subplot(gs[:,1],sharey=ax1)
    ax3 = fig.add_subplot(gs[:,2],sharey=ax1)

    text1(ax1)
    text2(ax2)
    text3(ax3)

    plt.tight_layout(pad=2,w_pad=2,h_pad=4)
    plt.suptitle('TEXT')
    plt.show()
```

```python
if __name__ == '__main__':
    plotText()
```
